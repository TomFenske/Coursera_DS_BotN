```python
#Dropping columns which we do not need
df_metro = df_metro.drop(['S-Bahnhof(ehem. Name)', 'Strecke', 'Linie(n)','Aufnahme', 'Einstellung','Bemerkungen', 'Unnamed: 6', 'Unnamed: 7','Bild'], axis=1)
df_metro.head()
```

```python
# Extracting the neighbourhoods and the Coordinates from the "Lage" column
df_metro['Neighbourhood']=df_metro['Lage'].str.extract('(\w*)', expand=True)
df_metro['Coordinates'] = df_metro['Lage'].str.extract('(\d.*)', expand=True)
df_metro.head()
```

```python
#Cleaning the Coordinates in order to express them as float later on
df_metro['Coordinates']=df_metro['Coordinates'].str.replace(' ', '')
df_metro['Coordinates']=df_metro['Coordinates'].str.replace('′', '\'')
df_metro['Coordinates']=df_metro['Coordinates'].str.replace('″', '\"')
df_metro.head()
```

```python
#Splitting Coordinates into Latitude and Longitude
df_metro[['Latitude','Longitude']]=df_metro['Coordinates'].str.split(",",expand=True)
df_metro.head()
```

```python
def coord_to_float(coord):
    degrees, minutes, seconds, direction = re.split('[°\'"]+', coord)
    latlon = float(degrees) + float(minutes)/60 + float(seconds)/(60*60);
    if direction in ('S','W'):
        latlon=latlon* -1
    return latlon


```
